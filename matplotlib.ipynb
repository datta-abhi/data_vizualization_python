```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
kohli_sharma_df = pd.read_csv('./datasets_for_viz/sharma-kohli.csv',dtype={'index':'string'})
```

```python
kohli_sharma_df.head(3)
```

```python
# plotting single batsmen stats
plt.plot(kohli_sharma_df['index'],kohli_sharma_df['V Kohli'])
```

```python
# plotting multiple graphs in same plot
# adding title and labels

plt.plot(kohli_sharma_df['index'],kohli_sharma_df['V Kohli'])
plt.plot(kohli_sharma_df['index'],kohli_sharma_df['RG Sharma'])

plt.title('Runs per season of R.Sharma and Virat')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

## Customizing color and line styles, data point marker

```python
# plotting multiple graphs in same plot
# adding title and labels

plt.plot(kohli_sharma_df['index'],kohli_sharma_df['V Kohli'],color = 'green',ls = 'dashed',lw = 4,marker = '*',label = 'Virat')
plt.plot(kohli_sharma_df['index'],kohli_sharma_df['RG Sharma'],color = 'black',ls = 'dashdot',lw = 3,marker = 's',label = 'Rohit')

plt.title('Runs per season of R.Sharma and Virat')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```
