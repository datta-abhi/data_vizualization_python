```python
kohli_sharma_df = pd.read_csv('./datasets_for_viz/sharma-kohli.csv',dtype={'index':'string'})
```

```python
# plotting single batsmen stats
plt.plot(kohli_sharma_df['index'],kohli_sharma_df['V Kohli'])
```
