```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
kohli_sharma_df = pd.read_csv('./datasets_for_viz/sharma-kohli.csv',dtype={'index':'string'})
```

```python
kohli_sharma_df.head(3)
```

```python
# plotting single batsmen stats
plt.plot(kohli_sharma_df['index'],kohli_sharma_df['V Kohli'])
```

```python
# plotting multiple graphs in same plot
# adding title and labels

plt.plot(kohli_sharma_df['index'],kohli_sharma_df['V Kohli'])
plt.plot(kohli_sharma_df['index'],kohli_sharma_df['RG Sharma'])

plt.title('Runs per season of R.Sharma and Virat')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

## Customizing color and line styles, data point marker

```python
# plotting multiple graphs in same plot
# adding title and labels

plt.plot(kohli_sharma_df['index'],kohli_sharma_df['V Kohli'],color = 'green',ls = 'dashed',lw = 4,marker = '*',label = 'Virat')
plt.plot(kohli_sharma_df['index'],kohli_sharma_df['RG Sharma'],color = 'black',ls = 'dashdot',lw = 3,marker = 's',label = 'Rohit')

plt.title('Runs per season of R.Sharma and Virat')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```

## Limiting axes of field of view --> done when theres outliers in the data

```python
# plotting multiple graphs in same plot
# adding title and labels

plt.plot(kohli_sharma_df['index'],kohli_sharma_df['V Kohli'],color = 'green',ls = 'dashed',lw = 4,marker = '*',label = 'Virat')
plt.plot(kohli_sharma_df['index'],kohli_sharma_df['RG Sharma'],color = 'black',ls = 'dashdot',lw = 3,marker = 's',label = 'Rohit')

plt.title('Runs per season of R.Sharma and Virat')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()

plt.xlim(['2009','2016'])
plt.ylim([0,1000])
```

# Scatterplot

```python
batter_df = pd.read_csv('./datasets_for_viz/batter.csv')
```

```python
batter_df.sample(3)
```

```python
batter_df.shape
```

```python
batter_df = batter_df[batter_df['runs']>2000]
```

```python
batter_df.shape
```

```python
plt.scatter(batter_df['avg'],batter_df['strike_rate'],marker = '*')
plt.title('Avg vs strike rate of batsmen in IPL')
plt.xlabel('Average')
plt.ylabel('Strike rate')
```

## size of points in scatterplot

```python
plt.scatter(batter_df['avg'],batter_df['strike_rate'],marker = 'o',s = batter_df['runs']/50)
plt.title('Avg vs strike rate of batsmen in IPL')
plt.xlabel('Average')
plt.ylabel('Strike rate')
```

# Barplot

# Multiple bar plots

```python
batsmen_record  = pd.read_csv('./datasets_for_viz/batsman_season_record.csv')
batsmen_record.head()
```

### We want to plot bars of each batsman and their runs in each season

```python
# shifting operation in x-axis
plt.bar(np.arange(batsmen_record.shape[0])-0.2,batsmen_record['2015'],width = 0.2,color = 'blue')
plt.bar(np.arange(batsmen_record.shape[0]),batsmen_record['2016'],width = 0.2,color = 'red')
plt.bar(np.arange(batsmen_record.shape[0])+0.2,batsmen_record['2017'],width = 0.2,color = 'green')

# changing the xlabels
plt.xticks(ticks=np.arange(batsmen_record.shape[0]),labels=batsmen_record['batsman'])
plt.show()
```
