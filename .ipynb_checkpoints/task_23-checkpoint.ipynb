### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
df = pd.read_csv('datasets_for_viz/polluted_cities.csv')
df.head(3)
```

```python
m1 = df['Country'].isin(['China','Iran'])
m
temp_df = df[m1].groupby('Year')['PM2.5'].sum()
plt.plot(temp_df)
plt.title('Yearwise PM2.5 Levels' )
plt.xlabel('Year')
plt.ylabel('PM2.5')
```

### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.hist(df['PM10'],density=True,bins = 50,edgecolor = 'black')

plt.title('Histogram of PM10')
plt.xlabel('PM10 Levels')
plt.ylabel('Probabinilty density')
plt.show()
```

### **`Problem-3:`** Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
p_mask = df['Country'] =='Poland'
c_mask = df['Country'] =='Chile'
plt.scatter(x = df[p_mask]['PM2.5'],y = df[p_mask]['PM10'],label = 'Poland',color = 'red',marker = 'D')
plt.scatter(x = df[c_mask]['PM2.5'],y = df[c_mask]['PM10'],label = 'Chile',color = 'Blue')

plt.title('PM2.5 vs PM10 for Poland and Chile')
plt.xlabel('PM2.5')
plt.ylabel('PM10')

plt.legend()
```

### **`Problem-4:`** Draw a pie chart of top 5 most frequent countries.

```python
plt.pie(df['Country'].value_counts().head(),labels= df['Country'].value_counts().head().index,autopct= '%.1f%%')

plt.show()
```

### **`Problem-5:`** Draw a bar chart which represents the counts of top 5 most frequent countries.



```python
plt.bar(df['Country'].value_counts().head().index,df['Country'].value_counts().head())

plt.show()
```

###`Problem-6`
Show a line plot of Total Profit for each month with below styling.
* Dotted Line
* Line Color Blue

```python
sales = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTJh6X4_mqixWsfK9mgkllGQkKYW9Wj9kOIMGY2uYsWeS8n5np87DO-SDGQWJ1HXEnxiOVFVzYFYEcR/pub?gid=558678488&single=true&output=csv')
sales.head(3)
```

###`Problem-7`
Show sales of each product in march month as pie chart.
* Show Percentage value
* Give Title "Sales in March"
* Explode ToothPaste with shadow

```python
march_sales = sales[sales['month_number']==3]
march_sales.drop(columns=['month_number','total_profit','total_units'],inplace=True)
march_sales.T.index
```

```python
march_sales.iloc[0]
```

```python
plt.pie(march_sales.iloc[0],labels=march_sales.iloc[0].index,autopct='%.0f%%',explode = [0,0,0.1,0,0,0],shadow = True)
plt.title('March Sales')
plt.show()
```
